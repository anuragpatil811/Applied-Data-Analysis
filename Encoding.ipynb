```python
da = df[df["species"]=="Adelie"]
dc = df[df["species"]=="Chinstrap"]
dg = df[df["species"]=="Gentoo"]
```

```python
sns.boxplot(da)
```

```python
da["flipper_length_mm"].describe()
```

```python
q1 = 186.0
q3 = 195.0
```

```python
IQR = q3 - q1 
IQR 
```

```python
uf = q3 + 1.5*IQR 
uf 
```

```python
lf = q1 - 1.5*IQR 
lf 
```

```python
da = da[(da["flipper_length_mm"]>lf)&(da["flipper_length_mm"]<uf)]
```

```python
sns.boxplot(data=da, y="flipper_length_mm")
```
