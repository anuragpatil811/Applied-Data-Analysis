```python
df1 = df[df["species"] == "Adelie"]
df2 = df[df["species"] == "Chinstrap"]
df3 = df[df["species"] == "Gentoo"]  
```

```python
df1.head()
```

```python
df1.describe()
```

```python
sns.boxplot(data=df1, y="flipper_length_mm")
```

```python
df1.describe()
```

```python
q3 = 195.0
q1 = 186.0
```

```python
IQR = q3 - q1
IQR
```

```python
uf = q3 + 1.5 * IQR
uf 
```

```python
lf = q1 - 1.5*IQR 
lf 
```

```python
df1 = df1[(df1["flipper_length_mm"]>lf) & (df1["flipper_length_mm"]<uf)]
df1 
```

```python
sns.boxplot(data=df1, y="flipper_length_mm")
```

```python
#Detect the outliers in body mass 
sns.boxplot(data=df1, y="body_mass_g")
```

```python
#CHINSTRAP SPECIES 
df2.info()
```

```python
sns.boxplot(data=df2, y="body_mass_g")
```

```python
df2.describe()
```

```python
#Find the first quartile and third quartile for body mass
q1 = 3487.50
q3 = 3950.00
```

```python
#Find Interquartile range for body mass
IQR = q3 - q1 
IQR 
```

```python
upper = q3 + 1.5 * IQR
upper 
```

```python
lower = q1 - 1.5*IQR
lower 
```

```python
df2 = df2[(df2["body_mass_g"]>lower) & (df2["body_mass_g"]<upper)]
df2 
```

```python
sns.boxplot(data=df2, y="body_mass_g")
```

```python
#GENTOO SPECIES
df3.info()
```

```python
sns.boxplot(data=df3, y="bill_length_mm")
```

```python
df3.describe()
```

```python
q1 = 45.30
q3 = 49.55
```

```python
#Interquartile Range of bill length of Gentoo species
IQR = q3 - q1 
IQR 
```

```python
upper = q3 + 1.5*IQR
upper 
```

```python
lower = q1 - 1.5*IQR 
lower 
```

```python
df3 =  df3[(df3["bill_length_mm"]>lower) & (df3["bill_length_mm"]<upper)]
df3 
```

```python
sns.boxplot(data=df3, y="bill_length_mm")
```

```python
import pandas as pd 
df = pd.concat([df1, df2, df3], ignore_index=True)
df 
```
