```python
import matplotlib.pyplot as plt
```

```python
years = [2015, 2016, 2017, 2018, 2019, 2020]
sales = [100, 150, 200, 250, 300, 400]
```

```python
plt.plot(years, sales, marker='o', linestyle='--', color='g', linewidth=2, label='Sales Trend')
plt.xlabel("Year")
plt.ylabel("Sales in Lakhs", fontsize=12, fontweight="bold")
plt.title("Company Sales Over years", fontsize=14, fontweight="bold")
plt.legend()
plt.show()
```
